# Data preparation

```python
# setup
import pandas as pd
```

## Data import

```python
URL = 'https://raw.githubusercontent.com/kirenz/datasets/master/3_4_data.csv'

df = pd.read_csv(URL)
```

```python
df
```

```python
df = df.melt(id_vars=['Month', 'Date', 'Goal'],
            value_vars=['Direct Sales', 'Indirect Sales'],
            var_name='Sales Type',
            value_name='Sales'
            )

df
```

### Create new variable

```python
df['Text'] = 'GOAL'
```

### Change data format

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.info()
```

```python
LIST_CAT = ['Month', 'Sales Type']

for i in LIST_CAT:
    df[i] = df[i].astype('category')
```

```python
df.info()
```

## Save data

```python
df.to_csv('3_4_data.csv', index=False)
```
